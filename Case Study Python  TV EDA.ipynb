##Data Cleaning

```python
## Fill null values Data Type String/Date with Unknown (views)
df_views['Channel2'] = df_views['Channel2'].fillna('Unknown')
df_views['RecordDate2'] = df_views['RecordDate2'].fillna('Unknown')
df_views['Duration 2'] = df_views['Duration 2'].fillna('Unknown')

## Fill null values Data Type String/Date with Unknown (profile)
df_profile['Name'] = df_profile['Name'].fillna('Unknown')
df_profile['Surname'] = df_profile['Surname'].fillna('Unknown')
df_profile['Gender'] = df_profile['Gender'].fillna('Unknown')
df_profile['Race'] = df_profile['Race'].fillna('Unknown')
df_profile['Province'] = df_profile['Province'].fillna('Unknown')
df_profile['Social Media Handle'] = df_profile['Social Media Handle'].fillna('Unknown')
df_profile["Email"] = df_profile["Email"].fillna('Unknown')
```

```python
## Fill null values Integer/Float with 0
df_profile['Age'] = df_profile['Age'].fillna(0)
df_profile['Age'] = df_profile['Age'].fillna(0)
```

```python
## Drop unnecessary columns
df_views = df_views.drop(columns=["Unnamed: 4"])
df_views = df_views.drop(columns=["Unnamed: 5"])
```

```python
df_views.columns
```

```python
#Upper case the first letter in Column called Race
df_profile['Race'] = df_profile['Race'].str.capitalize()

## Upper case the first letter in called Gender
df_profile['Gender'] = df_profile['Gender'].str.capitalize()
```

```python
df_views.head(10)
```

```python
df_views.info()
```

```python
## Merge the two df_views and df_profile DataFrames on UserID
df_views = df_views.merge(df_profile, on='UserID', how='left')
```

```python
df_views.head(10)
```

```python
# Replacing the original string column with SAST datetime
df_views['RecordDate2'] = pd.to_datetime(df_views['RecordDate2'], format='%m/%d/%y %H:%M')
df_views['RecordDate2'] = df_views['RecordDate2'] + pd.Timedelta(hours=2)

# Created Endtime by adding the Duration to RecordDate2
df_views['Endtime'] = df_views['RecordDate2'] + pd.to_timedelta(df_views['Duration 2'])

# Extract the Month, Day and Year.
df_views['Month'] = df_views['RecordDate2'].dt.month_name()
df_views['Day'] = df_views['RecordDate2'].dt.day_name()
df_views['Year'] = df_views['RecordDate2'].dt.year

# Extract the Hour and Minute.
df_views['Hour'] = df_views['RecordDate2'].dt.hour
df_views['Minute'] = df_views['RecordDate2'].dt.minute
```

### Data Visualization

```python
# Create a figure with 2 subplots (1 row, 2 columns)
fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# Bar graph Count by Gender (as pie chart)
No_of_Views_by_Gender = df_views.groupby(['Gender']).size().reset_index(name='Count')
No_of_Views_by_Gender = No_of_Views_by_Gender.sort_values(by='Count', ascending=False)

# Pie chart on first  ax[0]
ax[0].pie(No_of_Views_by_Gender['Count'], 
          labels=No_of_Views_by_Gender['Gender'], 
          autopct='%1.1f%%',
          colors=plt.cm.Set3(range(len(No_of_Views_by_Gender))),
          startangle=180)
ax[0].set_title('Views Distribution by Gender')
ax[0].axis('equal')  # Equal aspect ratio ensures the pie chart is circular

# Bar graph Count by Race
No_of_Views_by_Race = df_views.groupby(['Race']).size().reset_index(name='Count')
No_of_Views_by_Race = No_of_Views_by_Race.sort_values(by='Count', ascending=False)

# Bar chart second ax[1]
bars = ax[1].bar(No_of_Views_by_Race['Race'], No_of_Views_by_Race['Count'], color='skyblue')
ax[1].set_xlabel('Race')
ax[1].set_ylabel('Count')
ax[1].set_title('Number of Views by Race')

# Add value labels in the center of each bar using bar_label
ax[1].bar_label(bars, label_type='center', fontweight='bold')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()


```

```python
# Create a figure with 2 subplots (2 rows, 1 column)
fig, ax = plt.subplots(2, 1, figsize=(12, 12))

# Top 10 Most Viewed Channel
top_10_channels = df_views.groupby('Channel2')['UserID'].count().sort_values(ascending=False).head(10)

# Create the plot on ax[0]
top_10_channels.plot(kind='bar', color='skyblue', ax=ax[0], edgecolor='lightblue', label='Views')
ax[0].set_xlabel('Channels')
ax[0].set_ylabel('Number of Views')
ax[0].set_title('Top 10 Most Viewed Channels')
ax[0].legend()

# Add value labels in the center of each bar using bar_label
ax[0].bar_label(ax[0].containers[0], label_type='center', fontweight='bold')

# Top 10 Most Viewed Channel By Province
top_10_channels_by_province = df_views.groupby(['Province', 'Channel2'])['UserID'].count().sort_values(ascending=False).head(10)

# Create the plot on ax[1]
top_10_channels_by_province.plot(kind='bar', color='skyblue', ax=ax[1], edgecolor='lightblue', label='Views by Province')
ax[1].set_xlabel('Province and Channel')
ax[1].set_ylabel('Number of Views')
ax[1].set_title('Top 10 Most Viewed Channels by Province')
ax[1].legend()

# Add value labels in the center of each bar using bar_label
ax[1].bar_label(ax[1].containers[0], label_type='center', fontweight='bold')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Create a Age Classification.
conditions = [
    (df_views['Age'].between(0, 17)),
    (df_views['Age'].between(18, 24)),
    (df_views['Age'].between(25, 34)),
    (df_views['Age'].between(35, 44)),
    (df_views['Age'].between(45, 54)),
    (df_views['Age'] >= 55),
]

choices = ['Under 18', 'Young Adult', 'Adult', 'Mid-Aged-Adult', 'Mature Adult', 'Senior']

df_views['Age_Group'] = np.select(conditions, choices, default='Unknown')

df_age_group = df_views.groupby('Age_Group')['UserID'].count().to_frame().reset_index()
df_age_group = df_age_group.sort_values(by='UserID', ascending=False)
display(df_age_group)
```
